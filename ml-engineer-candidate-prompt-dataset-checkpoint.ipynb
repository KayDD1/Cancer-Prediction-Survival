```python
X = df_new.drop(['vital_status'], axis=1)
y = df_new['vital_status']
y = y.astype('int64')
```

```python
X.shape,y.shape
```

```python
norm = MinMaxScaler()
X_train_norm = norm.fit_transform(X)
X_train_norm
```

```python
pca = PCA(n_components=37)
pca.fit(X_train_norm)
X_train_norm_pca = pca.transform(X_train_norm)
```

```python
X_train_norm_pca.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_train_norm_pca,y, test_size=0.33, random_state=1)
print(X_train.shape,X_test.shape,y_train.shape, y_test.shape)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import SGDClassifier
```

```python
classifier_names = ['KNearest_Neighbors', 'SVM_Linear', 'SVM_Polynomial', 'SVM_RBF', 'Gaussian_Process', 'Gradient_Boosting_Machine',
                   'Decision_Tree', 'Extra_Tree', 'Random_Forest', 'Neural_Net', 'AdaBoost', 'Naive_Bayes', 'QDA', 'SGD']
classifiers = [
            KNeighborsClassifier(5),
            SVC(kernel='linear', C=0.025),
            SVC(kernel='poly', degree=3, C=0.025),
            SVC(kernel='rbf', C=1, gamma=2),
            GaussianProcessClassifier(1.0 * RBF(1.0)),
            GradientBoostingClassifier(n_estimators=100, learning_rate=1.0),
            DecisionTreeClassifier(max_depth=5),
            ExtraTreesClassifier(n_estimators=10, min_samples_split=2),
            RandomForestClassifier(max_depth=5, n_estimators=100),
            MLPClassifier(alpha=1, max_iter=1000),
            AdaBoostClassifier(n_estimators=100),
            GaussianNB(),
            QuadraticDiscriminantAnalysis(),
            SGDClassifier(loss='hinge', penalty='l2')]
```

```python
model_scores = []
for name, clf in zip(classifier_names, classifiers):
    clf.fit(X_train, y_train)
    model_score = clf.score(X_test, y_test) 
    model_scores.append(model_score)
model_scores    
```

```python
df_scores = pd.DataFrame()
df_scores['classifier_name'] = classifier_names
df_scores['model_score'] = model_scores
```

```python
xy = sns.light_palette('blue', as_cmap=True)
s = df_scores.style.background_gradient(cmap=xy)
s
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.metrics import r2_score
```

```python
mlp = MLPClassifier(max_iter=500)
```

```python
param_space = {
    'hidden_layer_sizes': [(150,150,150), (150,200,150), (200,)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001,0.05],
    'learning_rate': ['constant', 'adaptive'],
}
```

```python
clf = GridSearchCV(mlp, param_space, n_jobs=-1, cv=3)
clf.fit(X_train, y_train)
```

```python
#Best Parameter set
print('Best Parameters found: \n', clf.best_params_)
# Results
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print('%.3f (+/-%.03f) for %r' % (mean, std * 2, params))
```

```python
y_actual, y_pred = y_test, clf.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print('Results on the test set for MULTILAYER PERCEPTRON: ')
print(classification_report(y_actual, y_pred))
report = classification_report(y_actual, y_pred, output_dict=True)
df_report = pd.DataFrame(report)
```
